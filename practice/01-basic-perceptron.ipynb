```python
# Step 01: Importing using libraries

from tensorflow import keras
import numpy as np
from keras import models
from keras.models import Sequential
from keras.layers.core import Dense
from keras import layers
from keras import optimizers
import matplotlib.pyplot as plt
```

```python
# Step 02: Creating and Checking Data

celcius_degree= np.array([-40, -10, 0, 8, 15, 22, 38], dtype= 'float')
fahrenheit_degree= np.array([-40, 14, 32, 46, 59, 72, 100], dtype= 'float')

for f,c in enumerate(celcius_degree):
    print("{} degree(s) Celcius = {} degree(s) Fahrenheit.".format(c, fahrenheit_degree[f]))
```

```python
print(celcius_degree*9/5 + 32)
```

##### Celcius to Fahrenheit & F to C conversion formulas:

##### It shows that weight value (w) should converges to 0.55 and 1.8 respectively and bias value (b) should converge 32.

```python
# Step 03: Building the Model

def TrainValModel():
    model= keras.Sequential()
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer= keras.optimizers.Adam(learning_rate=0.1))
    model.fit(celcius_degree, fahrenheit_degree, epochs= 500, verbose= True)
    
    ValidateModel(model, celcius_degree, fahrenheit_degree)
    print('Model training is finished!\n')
    return model
```

##### Building network can be presented below:


```python
#Step 03 (cont.): Validation function of the model

def ValidateModel(model,celcius_degree, fahrenheit_degree):
    predictions= model.predict(celcius_degree)
    
    for idx in range(len(predictions)):
        print(f'{celcius_degree[idx]} C degree -> Predicted Fahrenheit value & Actual Fahrenheit value: {predictions[idx]} F & Real F: {fahrenheit_degree[idx]} F\n')
```

```python
# Step 04: Training the Model

model= TrainValModel()
history= model.fit(celcius_degree, fahrenheit_degree, epochs= 500, verbose= True)
```

```python
# Step 05 (cont.): Creating a function to make predictions

def PredictCelcius(model, celcius_degree):
    return model.predict(np.array([celcius_degree]))
```

```python
# Step 05 : Run the Model

prediction_celcius= float(input('Enter the C value for F prediction: '))
print(PredictCelcius(model, prediction_celcius))
```

```python
# Step 06: Plotting

plt.plot(history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs')
```
